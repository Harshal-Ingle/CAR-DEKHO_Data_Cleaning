# car_name

```python
# x = df['car_name'][0]
# x.index(" ")               # To find the index number   
# x[:x.index(" ")]
```

```python
def company_name(x):
    return x[:x.index(" ")]
```

```python
df['company_name'] = df['car_name'].apply(company_name)
```

```python
df.head()
```

```python
def car_name(x):
    return x[x.index(" ")+1:]
```

```python
df['car_name'] = df['car_name'].apply(car_name)
```

```python
df
```

# car_prices_in_rupee

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].str.replace(",","")
```

```python
# Problems = 35,000, 38.90 Lakh, 1.02 Crore
```

```python
def rupee_change(x):
    p = x.split(" ")
    try:
        if p[1] == 'Lakh':
            return str(round(float(p[0])*100000,1))
        elif p[1] == 'Crore':
            return str(round(float(p[0])*10000000,1))
    except:
            return x
                       
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].apply(rupee_change)
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].astype('float64')
```

```python
df
```

# kms_driven

```python
df['kms_driven'] = df['kms_driven'].str.replace(",", "")
df['kms_driven'] = df['kms_driven'].str.replace(" kms","")
```

```python
df['kms_driven'] = df['kms_driven'].astype("int64")
```

```python
df
```

# engine

```python
df['engine'] = df['engine'].str.replace("cc","")
```

```python
df['engine'] = df['engine'].astype("int64")
```

```python
df.info()
```

# Seats

```python
df['Seats'] = df['Seats'].str.replace(" Seats", "")
df['Seats'] = df['Seats'].astype("int64")
```
